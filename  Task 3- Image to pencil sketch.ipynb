# IMAGE TO PENCIL SKETCH WITH PYTHON

# Importing Modules

```python
import cv2
import matplotlib.pyplot as plt
```

# Displaying the image

```python
#Reading the image
img = cv2.imread("WallpaperCat.jpg")
```

```python
#Showing the image
image1 = plt.imshow(img[...,::-1])
print(image1)
```

# Converting Image to Grescale

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray_img,cmap='gray')
```

# Inverting this Grayscale Image

```python
img2 = cv2.bitwise_not(gray_img)
```

```python
plt.imshow(img2,cmap='gray')
```

# Blurring the image

```python
img3 = cv2.GaussianBlur(img2,(33,33),0)
```

```python
plt.imshow(img3,cmap='gray')
```

# Sketching the final Image

```python
img4 = cv2.divide(img2,img3, scale=255.0)
```

```python
plt.imshow(img4,cmap='gray')
```

```python
cv2.imwrite("Final_sketch.png",img4)
```
